```python
import numpy as nd
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nacimientos=pd.DataFrame({'municipio':['Mty','Apodaca','Escobedo','Santiago'],'bb_nacidos':[233,569,780,123],'Moridos':[123,80,23,12]})
```

```python
nacimientos=nacimientos.sort_values('bb_nacidos',ascending=False)
```

```python
plt.bar(nacimientos['municipio'],nacimientos['bb_nacidos'])
plt.xlabel('Municipios')
plt.ylabel('Cantidad de bebes')
plt.title('Nacimientos en N.L en 2020')
plt.show()
```

```python
personas=['Juan','Pedro','Ana','Ale']
edades=[23,18,20,31]
```

```python
plt.bar(personas,edades)
plt.show()
```
