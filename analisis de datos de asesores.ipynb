```python
# librerias que utilizaremos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statistics as stl
import scipy.stats as st
```

# Datos con lo que trabajamos

```python
# lectura de los datos
datos=pd.read_excel('Info Asesores.xlsx')
datos=datos.set_index('Id')
```

```python
# Vista previa de los datos
datos
```

# Estadistica Descriptiva Básica

```python
promedio=[
    np.mean(datos['Estatura']),
    np.mean(datos['Peso']),
    np.mean(datos['Horas estudio (sem)']),
    np.mean(datos['Horas recreación (P)(SEM)']),
    np.mean(datos['Horas recreación (A)(SEM)']),
    np.mean(datos['Horas sueño (D)']),
    np.mean(datos['Traslado Casa-Uni (min)'])]
```

```python
medias=[
np.median(datos['Estatura']),
np.median(datos['Peso']),
np.median(datos['Horas estudio (sem)']),
np.median(datos['Horas recreación (P)(SEM)']),
np.median(datos['Horas recreación (A)(SEM)']),
np.median(datos['Horas sueño (D)']),
np.median(datos['Traslado Casa-Uni (min)'])]
```

```python
moda=[
stl.mode(datos['Estatura']),
stl.mode(datos['Peso']),
stl.mode(datos['Horas estudio (sem)']),
stl.mode(datos['Horas recreación (P)(SEM)']),
stl.mode(datos['Horas recreación (A)(SEM)']),
stl.mode(datos['Horas sueño (D)']),
stl.mode(datos['Traslado Casa-Uni (min)'])]
```

```python
minimo=[
datos['Estatura'].min(),
datos['Peso'].min(),
datos['Horas estudio (sem)'].min(),
datos['Horas recreación (P)(SEM)'].min(),
datos['Horas recreación (A)(SEM)'].min(),
datos['Horas sueño (D)'].min(),
datos['Traslado Casa-Uni (min)'].min()
]
```

```python
maximo=[
datos['Estatura'].max(),
datos['Peso'].max(),
datos['Horas estudio (sem)'].max(),
datos['Horas recreación (P)(SEM)'].max(),
datos['Horas recreación (A)(SEM)'].max(),
datos['Horas sueño (D)'].max(),
datos['Traslado Casa-Uni (min)'].max()
]
```

```python
rango=[
    datos['Estatura'].max()-datos['Estatura'].min(),
    datos['Peso'].max()-datos['Peso'].min(),
    datos['Horas estudio (sem)'].max()-datos['Horas estudio (sem)'].min(),
    datos['Horas recreación (P)(SEM)'].max()-datos['Horas recreación (P)(SEM)'].min(),
    datos['Horas recreación (A)(SEM)'].max()-datos['Horas recreación (A)(SEM)'].min(),
    datos['Horas sueño (D)'].max()-datos['Horas sueño (D)'].min(),
    datos['Traslado Casa-Uni (min)'].max()-datos['Traslado Casa-Uni (min)'].min()
]
```

```python
desviacion=[
    np.std(datos['Estatura']),
    np.std(datos['Peso']),
    np.std(datos['Horas estudio (sem)']),
    np.std(datos['Horas recreación (P)(SEM)']),
    np.std(datos['Horas recreación (A)(SEM)']),
    np.std(datos['Horas sueño (D)']),
    np.std(datos['Traslado Casa-Uni (min)']),
    ]
```

```python
p_cuantil=[
    datos['Estatura'].quantile(0.25),
    datos['Peso'].quantile(0.25),
    datos['Horas estudio (sem)'].quantile(0.25),
    datos['Horas recreación (P)(SEM)'].quantile(0.25),
    datos['Horas recreación (A)(SEM)'].quantile(0.25),
    datos['Horas sueño (D)'].quantile(0.25),
    datos['Traslado Casa-Uni (min)'].quantile(0.25),
]
```

```python
s_cuantil=[
    datos['Estatura'].quantile(0.5),
    datos['Peso'].quantile(0.5),
    datos['Horas estudio (sem)'].quantile(0.5),
    datos['Horas recreación (P)(SEM)'].quantile(0.5),
    datos['Horas recreación (A)(SEM)'].quantile(0.5),
    datos['Horas sueño (D)'].quantile(0.5),
    datos['Traslado Casa-Uni (min)'].quantile(0.5),
]
```

```python
t_cuantil=[
    datos['Estatura'].quantile(0.75),
    datos['Peso'].quantile(0.75),
    datos['Horas estudio (sem)'].quantile(0.75),
    datos['Horas recreación (P)(SEM)'].quantile(0.75),
    datos['Horas recreación (A)(SEM)'].quantile(0.75),
    datos['Horas sueño (D)'].quantile(0.75),
    datos['Traslado Casa-Uni (min)'].quantile(0.75),
]
```

```python
titulos=['Promedio','Medias','Moda','Rango','Desviacion','Minimo','P_cuantil','S_cuantil','T_cuantil','Maximo']
```

```python
e=[promedio,medias,moda,rango,desviacion,minimo,p_cuantil,s_cuantil,t_cuantil,maximo]
```

```python
Descriptivas=pd.DataFrame(e,columns=['Estatura','Peso','Horas estudio (sem)','Horas recreación (P)(SEM)','Horas recreación (A)(SEM)','Horas sueño (D)','Traslado Casa-Uni (min)'])
```

```python
# Estadistica descriptiva de los datos
Descriptivas['Titulo']=titulos
Descriptivas=Descriptivas.set_index('Titulo')
Descriptivas
```

# Visualización de Datos

```python
# Grafica de las estauras
plt.hist(datos['Estatura'], color="skyblue")
plt.xlabel('Estaura del asesor')
plt.ylabel('Cantidad de asesores')
plt.title('Cantidad de asesores por estatura')
plt.show()
```

```python
#Prueba de kolgomorov-Smirnof
Normales=(datos['Estatura']-Descriptivas.iloc[0,0])/Descriptivas.iloc[4,0]
k_statistic, p_value = st.kstest(Normales, 'norm')
print("Estadístico de la prueba K-S:", k_statistic)
print("Valor p:", p_value)
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Se rechaza la hipótesis nula. La muestra no sigue una distribución normal.")
else:
    print("No se rechaza la hipótesis nula. La muestra sigue una distribución normal.")
```

```python
# calculo de la varianza
f=(datos['Estatura']-datos['Estatura'].mean())**2
f.sum()/(datos['Estatura'].count())
```

# La distrubición de las estaturas en una normal(1.67,0.0082)

```python
# Grafica del peso
plt.hist(datos['Peso'], color='Skyblue')
plt.xlabel('Peso de los asesores')
plt.ylabel('Cantidad de asesores')
plt.title('Cantidad de asesores por peso')
plt.show()
```

```python
#Prueba de kolgomorov-Smirnof
Normales=(datos['Peso']-Descriptivas.iloc[0,1])/Descriptivas.iloc[4,1]
k_statistic, p_value = st.kstest(Normales, 'norm')
print("Estadístico de la prueba K-S:", k_statistic)
print("Valor p:", p_value)
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Se rechaza la hipótesis nula. La muestra no sigue una distribución normal.")
else:
    print("No se rechaza la hipótesis nula. La muestra sigue una distribución normal.")
```

```python
# Calculo de la varianza
f=(datos['Peso']-datos['Peso'].mean())**2
f.sum()/(datos['Peso'].count())
```

# La distribución del peso es una normal(71.86,416.64)
