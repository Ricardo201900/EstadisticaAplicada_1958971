```python
renunciar=pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/desercion_laboral.csv')
```

```python
renunciar.isnull().sum()
```

```python
renunciar=renunciar.dropna()
```

```python
renunciar.duplicated().sum()
```

```python
renunciar=renunciar.drop_duplicates(keep='first').reset_index(drop=True)
```

```python
renunciar.dtypes
```

```python
renunciar['satisfaccion']=renunciar['satisfaccion'].astype('int64')
renunciar['evaluacion']=renunciar['evaluacion'].astype('int64')
```

```python
dummies1 = pd.get_dummies(renunciar['departamento'], prefix='departamento').astype(int)
dummies2 = pd.get_dummies(renunciar['salario'], prefix='salario').astype(int)
```

```python
dummies2
```

```python
dummies1=dummies1[['departamento_administración','departamento_contabilidad','departamento_desarrollo de producto','departamento_informática','departamento_mantenimiento','departamento_mercadotecnia','departamento_recursos humanos','departamento_soporte','departamento_ventas']]
dummies2=dummies2[['salario_bajo','salario_medio']]
```

```python
renunciar=pd.concat([renunciar,dummies1,dummies2],axis=1)
```

```python
renunciar=renunciar.drop('departamento',axis=1)
renunciar=renunciar.drop('salario',axis=1)
```

```python
renunciar.dtypes
```

```python
renunciar.isnull().sum()
```

```python
renunciar.describe()
```

```python
cph1 = CoxPHFitter().fit(renunciar, 'semestre', 'renunciar')
cph1.print_summary()
cph1.params_
cph1.hazard_ratios_
```
